Import the necessary libraries.

```python
import chess
import chess.pgn
import chess.svg
from IPython.display import display, SVG, HTML
from io import StringIO 
import subprocess
import random
from gifpgn import CreateGifFromPGN
from IPython.display import Image
```

Now, define some helper functions.
- **get_svg**: get the end-board of some pgn-file as an svg-image
- **run_swipl**: run our program on some pgn-file to get the next pgn-state as a string
- **write_pgn**: write out a pgn-string to tpm.pgn

An example of the parameter file_location could be "tests/examples/london.pgn".

```python
def get_svg(file_location, highlight=False):
    with open(file_location, "r") as f:
        pgn = f.read()
    game = chess.pgn.read_game(StringIO(pgn))
    board = game.end().board()
    if highlight:
        last_move = board.peek()
        return chess.svg.board(board, lastmove=last_move)
    else:
        return chess.svg.board(board)
```

```python
def run_swipl(file_location):
    command = f"swipl -t halt -f -q -O ./src/main.pl -- {file_location}"
    process = subprocess.Popen(command.split(), stdout=subprocess.PIPE, stderr=subprocess.PIPE) 
    out, err = process.communicate()
    result = out.decode("utf-8")
    return result
```

```python
def write_pgn(pgn_string, koth=False):
    with open("presentatie-files/tmp.pgn", 'w') as f:
        if koth:
            f.write('[Rules "koth"]' + '\n' + pgn_string)
        else:
            f.write(pgn_string)
```

The main function **show_result** will show the original board on the left and the result of our program on the right.

```python
def show_result(file_location):
    svg_board1 = get_svg(file_location)
    result = run_swipl(file_location)
    write_pgn(result)
    svg_board2 = get_svg("presentatie-files/tmp.pgn", highlight=True)

    combined_html = f"""
    <div style="display: flex; justify-content: space-around;">
        {svg_board1}
        {svg_board2}
    </div>
    """
    
    display(HTML(combined_html))
```

### Vs. random

```python
random.seed(2)
def create_gif_vs_random(koth=False):
    # initial move
    board = chess.Board()
    board.push(random.choice(list(board.legal_moves)))
    pgn = str(chess.pgn.Game.from_board(board))
    write_pgn(pgn, koth=koth)
    
    for i in range(2, 40):
        if i % 2 == 1:
            # own engine
            pgn = run_swipl("presentatie-files/tmp.pgn")
            game = chess.pgn.read_game(StringIO(pgn))
            board = game.end().board()
        else:
            # random move
            with open("presentatie-files/tmp.pgn", "r") as f:
                pgn = f.read()
            game = chess.pgn.read_game(StringIO(pgn))
            board = game.end().board()
            board.push(random.choice(list(board.legal_moves)))
        write_pgn(str(chess.pgn.Game.from_board(board)), koth=koth)
    
    with open("presentatie-files/tmp.pgn", "r") as f:
        pgn = f.read()
    gif = CreateGifFromPGN(pgn)
    gif.enable_arrows = True
    gif.generate("presentatie-files/generated_gif.gif")
    
```
